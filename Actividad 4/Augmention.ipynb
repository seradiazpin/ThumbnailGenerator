```python
import Augmentor
from PIL import Image
import os
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid


PATH = "..\\..\\ThumbnailGenerator\\Actividad 4\\TestImage\\"

def plot_img(imgs):
    
    fig = plt.figure(figsize=(30., 30.))
    grid = ImageGrid(fig, 111,  
                     nrows_ncols=(3, 4), 
                     axes_pad=0.1, 
                     )

    for ax, im in zip(grid, imgs):
        # Iterating over the grid returns the Axes.
        ax.imshow(im)
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.set_xticks([])
        ax.set_yticks([])

def reset_pipeline_flip(path, flip=True):
    p = Augmentor.Pipeline(path)
    if flip:
        p.flip_left_right(probability=1)
    return p

def augmentation_examples(PATH):
        
    # More brigthness
    p = reset_pipeline_flip(PATH,False)
    p.random_brightness(probability=1.0, min_factor=0.5, max_factor=1.0)
    p.sample(1)
    
    # less brigthness
    p = reset_pipeline_flip(PATH,False)
    p.random_brightness(probability=1.0, min_factor=0.5, max_factor=1.0)
    p.sample(1)

    #  random_color
    p = reset_pipeline_flip(PATH,False)
    p.random_color(probability=1.0, min_factor=0.3, max_factor=0.6)
    p.sample(1)
    
    # random_contrast
    p = reset_pipeline_flip(PATH,False)
    p.random_contrast(probability=1.0, min_factor=0.2, max_factor=0.5)
    p.sample(1)

    # 4 - random_distortion
    p = reset_pipeline_flip(PATH,False)
    p.random_distortion(probability=1, grid_width=16, grid_height=16, magnitude=8)
    p.sample(1)
    
    # 5 - crop_random 90%
    p = reset_pipeline_flip(PATH,False)
    p.crop_random(probability=1, percentage_area=0.9)
    p.resize(probability=1.0, width=1920, height=1080)
    p.sample(1)
    
    p = reset_pipeline_flip(PATH,False)
    p.shear(probability=1, max_shear_left = 5, max_shear_right = 5)
    p.sample(1)
    p.sample(1)
    p.sample(1)
    
    p = reset_pipeline_flip(PATH,False)
    p.skew(probability=1, magnitude=0.3)
    p.sample(1)
    p.sample(1)
    p.sample(1)
    
def augmentation_examples_flip(PATH):
    # Flip
    p = reset_pipeline_flip(PATH)
    p.sample(1)
    
    # More brigthness
    p = reset_pipeline_flip(PATH)
    p.random_brightness(probability=1.0, min_factor=0.5, max_factor=1.0)
    p.sample(1)
    
    # less brigthness
    p = reset_pipeline_flip(PATH)
    p.random_brightness(probability=1.0, min_factor=0.5, max_factor=1.0)
    p.sample(1)

    # random_color
    p = reset_pipeline_flip(PATH)
    p.random_color(probability=1.0, min_factor=0.3, max_factor=0.6)
    p.sample(1)

    # random_contrast
    p = reset_pipeline_flip(PATH)
    p.random_contrast(probability=1.0, min_factor=0.2, max_factor=0.5)
    p.sample(1)

    # random_distortion
    p = reset_pipeline_flip(PATH)
    p.random_distortion(probability=1, grid_width=16, grid_height=16, magnitude=8)
    p.sample(1)
    
    # crop_random 90%
    p = reset_pipeline_flip(PATH)
    p.crop_random(probability=1, percentage_area=0.9)
    p.resize(probability=1.0, width=1920, height=1080)
    p.sample(1)
    
    p = reset_pipeline_flip(PATH)
    p.shear(probability=1, max_shear_left = 5, max_shear_right = 5)
    p.sample(1)
    p.sample(1)
    p.sample(1)
    
    p = reset_pipeline_flip(PATH)
    p.skew(probability=1, magnitude=0.3)
    p.sample(1)
    p.sample(1)
    p.sample(1)

    
```
