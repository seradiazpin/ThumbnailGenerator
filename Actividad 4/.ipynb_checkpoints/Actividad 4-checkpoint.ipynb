```python
import numpy as np
import imgaug.augmenters as iaa
import cv2
import glob
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid

fps = glob.glob("..\\..\\Imagenes_Limpias\\Agua_2.jpg")
images = [cv2.imread(fp) for fp in fps]
images_aug = {"original": images[0]}

def image_augmentation(images, augmentation_op):
    seq = iaa.Sequential(augmentation_op)
    return seq(images=images)

def plot_img(imgs):
    fig = plt.figure(figsize=(20., 20.))
    grid = ImageGrid(fig, 111,  
                     nrows_ncols=(4, 3), 
                     axes_pad=0.3, 
                     )

    for ax, im in zip(grid, imgs):
        ax.set_title(im)
        # Iterating over the grid returns the Axes.
        ax.imshow(imgs[im])
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
import os
import numpy as np
import imgaug.augmenters as iaa
import cv2
import glob

def image_augmentation(images, augmentation_op):
    seq = iaa.Sequential(augmentation_op)
    return seq(images=images)

def Augmentate(operations, img_class, destination):
    fps = glob.glob("..\\..\\Imagenes_Limpias\\{0}*.jpg".format(img_class))
    if not os.path.exists(destination):
        os.makedirs(destination)
    images = [cv2.imread(fp) for fp in fps]
    for operation in operations:
        print("Start {0}".format(operation))
        images_aug = image_augmentation(images, operations[operation])
        for i in range(len(images_aug)):
            cv2.imwrite(os.path.join(destination,"{0}_{1}_{2}.jpg".format(img_class,str(operation),i)), images_aug[i])
        print("End {0}".format(operation))
```
