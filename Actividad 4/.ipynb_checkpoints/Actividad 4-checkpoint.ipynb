```python
import numpy as np
import imgaug.augmenters as iaa
import cv2
import glob
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid

fps = glob.glob("..\\Imagenes_Limpias\\Desierto_2.jpg")
images = [cv2.imread(fp) for fp in fps]
images_aug = [cv2.imread(fp) for fp in fps]

def image_augmentation(images, augmentation_op):
    seq = iaa.Sequential(augmentation_op)
    return seq(images=images)

def plot_img(imgs):
    
    fig = plt.figure(figsize=(20., 20.))
    grid = ImageGrid(fig, 111,  
                     nrows_ncols=(5, 3), 
                     axes_pad=0.1, 
                     )

    for ax, im in zip(grid, imgs):
        # Iterating over the grid returns the Axes.
        ax.imshow(im)
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
operation = [
iaa.FastSnowyLandscape(
lightness_threshold=50,
lightness_multiplier=2.5)
]

images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation = [iaa.Clouds()]

images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation =[iaa.Fog()]

images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation =[iaa.Snowflakes(flake_size=(0.7, 0.95), speed=(0.001, 0.03))]
images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation = [iaa.Rain(drop_size=(0.10, 0.20))]
images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation = [iaa.Fliplr(1)]
images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation = [iaa.PerspectiveTransform(scale=(0.01, 0.09))]
images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation = [iaa.ShearX((-20, 20))]
images_aug.append(image_augmentation(images, operation)[0])
```

```python
operation = [iaa.Rain(drop_size=(0.10, 0.20)),iaa.Fliplr(1)]
images_aug.append(image_augmentation(images, operation)[0])
```

```python
plot_img(images_aug)
```

```python
cv2.imwrite('new.jpg', images_aug[1])
```
