```python
from PIL import Image, ImageStat
import os, random
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid

columns_names =["number","File Name", "Class", "Size", "Width", "Height", "Extrem Values", "Number Pixels", "Sum of all pixels", "Squared sum", "Mean", "Root-mean-square", "Variance", "Standar deviation"]
images_description = []
def image_descriptors(folder):
    files = os.listdir(folder)
    for i in range(len(files)):
        name_parts = files[i].split("_")
        img = Image.open(os.path.join(folder,files[i]))
        width, height = img.size
        stat = ImageStat.Stat(img)
        image_data = [name_parts[1],files[i],name_parts[0], os.path.getsize(os.path.join(folder,files[i])),width,height, stat.extrema[0], stat.count[0], stat.sum[0], stat.sum2[0], stat.mean[0], stat.rms[0], stat.var[0], stat.stddev[0]]
        images_description.append(image_data)
    
    description = pd.DataFrame(images_description, columns =columns_names)
    return description

def plot_img(imgs):
    
    fig = plt.figure(figsize=(10., 10.))
    grid = ImageGrid(fig, 111,  
                     nrows_ncols=(3, 3), 
                     axes_pad=0.1, 
                     )

    for ax, im in zip(grid, imgs):
        # Iterating over the grid returns the Axes.
        ax.imshow(im)
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.set_xticks([])
        ax.set_yticks([])
    
```

```python
DESTINATION = "..\\Imagenes_Limpias\\"
description = image_descriptors(DESTINATION)
description.head()
```

```python
compression_opts = dict(method='zip',archive_name='description_images.csv')  
description.to_csv('out.zip', index=False,compression=compression_opts)  
```

```python
DESTINATION = "..\\Imagenes_Limpias\\"
imgs = []
for i in range(10):
    imgs.append(Image.open(os.path.join(DESTINATION,random.choice(os.listdir(DESTINATION)))))

plot_img(imgs)
```
