# Images to JPG

```python
from PIL import Image
import os

NAME_STANDAR = "{TYPE}_{NUMBER}.jpg"

def PNG2JPG(source, destination):
    img = Image.open(source).convert("RGB")
    img.save(destination, "jpeg")
    
def convert_png_in_folder(folder, category, destination):
    if not os.path.exists(destination):
        os.makedirs(destination)
    files = os.listdir(folder)
    for i in range(len(files)):
        new_name = NAME_STANDAR.format(TYPE = category, NUMBER= i+1 )
        print(os.path.join(folder,files[i]),"----------->", os.path.join(destination,new_name))
        PNG2JPG(os.path.join(folder,files[i]), os.path.join(destination,new_name))
```

## Agua

```python
SOURCE = "..\\Imagenes_Tesis\\Agua\\"
DESTINATION = "..\\Imagenes_Limpias\\"

convert_png_in_folder(SOURCE, "Agua", DESTINATION)
```
